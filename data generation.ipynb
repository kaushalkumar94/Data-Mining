```python
import pandas as pd
import numpy as np
import json
import random

# Set seed for reproducibility
np.random.seed(42)

# 1. Generate Transactions CSV (250 rows)
cust_ids = [f'C{str(i).zfill(3)}' for i in range(1, 251)]
transactions = {
    'CustomerID': [random.choice(cust_ids) for _ in range(250)],
    'Monthly_Spend': [round(random.uniform(10.0, 5000.0), 2) for _ in range(250)],
    'Account_Balance': [round(random.uniform(0.0, 20000.0), 2) for _ in range(250)],
    'Trans_Count': [random.randint(1, 100) for _ in range(250)]
}

df_trans = pd.DataFrame(transactions)

# --- Add Messiness to CSV ---
# Add 10% duplicates
duplicates = df_trans.sample(25)
df_trans = pd.concat([df_trans, duplicates], ignore_index=True)

# Add missing values (NaN)
df_trans.loc[df_trans.sample(15).index, 'Monthly_Spend'] = np.nan
df_trans.loc[df_trans.sample(10).index, 'Account_Balance'] = np.nan

df_trans.to_csv('transactions_dirty.csv', index=False)

# 2. Generate Customers JSON (250 records)
genders = ["M", "F", "male", "female", "Mle", "FEMALE", "m"]
regions = ["North", "South", "East", "West", "north", "south", "EAST"]

customer_data = []
for cid in cust_ids:
    customer_data.append({
        "CustomerID": cid,
        "Age": random.randint(18, 90) if random.random() > 0.05 else random.choice([150, -5, None]),
        "Gender": random.choice(genders),
        "Credit_Score": random.randint(300, 850),
        "Region": random.choice(regions)
    })

with open('customers_dirty.json', 'w') as f:
    json.dump(customer_data, f, indent=4)

print("Files 'transactions_dirty.csv' and 'customers_dirty.json' have been created!")
```
