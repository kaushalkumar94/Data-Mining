```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd     # ADD THIS
 
# 1. Generate Dataset
np.random.seed(42)

n_students = 50

# Mathematics -> Normal Distribution
math_marks = np.random.normal(loc=70, scale=10, size=n_students)

# Physics -> Uniform Distribution
physics_marks = np.random.uniform(low=50, high=90, size=n_students)

# Programming -> Gamma Distribution
programming_marks = np.random.gamma(shape=5, scale=10, size=n_students)

# Clip marks
math_marks = np.clip(math_marks,0,100)
physics_marks = np.clip(physics_marks,0,100)
programming_marks = np.clip(programming_marks,0,100)

 
```

```python
# 2. Calculate Average + Grades
average_marks = (
    math_marks +
    physics_marks +
    programming_marks
)/3


grades = []

for avg in average_marks:

    if avg >= 80:
        grades.append('A')

    elif avg >= 65:
        grades.append('B')

    elif avg >= 50:
        grades.append('C')

    else:
        grades.append('D')
```

```python
# . Create Dataset Table
df = pd.DataFrame({

    "Student_ID": np.arange(1, n_students+1),

    "Mathematics": np.round(math_marks,2),

    "Physics": np.round(physics_marks,2),

    "Programming": np.round(programming_marks,2),

    "Average": np.round(average_marks,2),

    "Grade": grades
})
```

```python
 # Save Dataset
df.to_csv("student_marks_dataset.csv", index=False)

print("Dataset saved successfully as student_marks_dataset.csv")
print(df.head())
```

```python
# 3. Grade Classification
grades = []

for avg in average_marks:

    if avg >= 80:
        grades.append('A')

    elif avg >= 65:
        grades.append('B')

    elif avg >= 50:
        grades.append('C')

    else:
        grades.append('D')
```

```python
# Grade Count
grade_labels = ['A','B','C','D']
grade_counts = [grades.count(g) for g in grade_labels]

# 4. Average Subject Marks

avg_math = np.mean(math_marks)
avg_physics = np.mean(physics_marks)
avg_programming = np.mean(programming_marks)

subject_names = ['Mathematics','Physics','Programming']
subject_averages = [
    avg_math,
    avg_physics,
    avg_programming
]
```

```python
# 5. Create Subplots
plt.figure(figsize=(16,10))
 
```

```python
# BAR CHART (Average Marks)
plt.subplot(2,3,1)

plt.bar(subject_names,subject_averages)

plt.title("Average Marks Across Subjects")

plt.xlabel("Subjects")
plt.ylabel("Average Marks")
```

```python
# HISTOGRAM (Programming Marks)
plt.subplot(2,3,2)

plt.hist(programming_marks,
         bins=10,
         edgecolor='black')

plt.title("Distribution of Programming Marks")

plt.xlabel("Marks")
plt.ylabel("Number of Students")
```

```python
# PIE CHART (Grade Distribution)
plt.subplot(2,3,3)

plt.pie(
    grade_counts,
    labels=grade_labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Grade Distribution")
```

```python
# SCATTER PLOT
plt.subplot(2,3,4)

plt.scatter(
    math_marks,
    physics_marks
)

plt.title("Correlation Between Mathematics and Physics")

plt.xlabel("Mathematics Marks")
plt.ylabel("Physics Marks")
```

```python
# BOX PLOT
plt.subplot(2,3,5)

plt.boxplot(
    [math_marks,
     physics_marks,
     programming_marks],

    labels=subject_names
)

plt.title("Spread and Variability Across Subjects")

plt.ylabel("Marks")


# Layout Adjustment
plt.tight_layout()

plt.show()
```
